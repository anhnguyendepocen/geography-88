# Simple web maps in python, using folium
First import `folium` and run its `initialize_notebook()` command.

```python
import folium as fm
fm.initialize_notebook()
```

```python
my_map = fm.Map(location=[37.86, -122.25], zoom_start=12)
my_map
```

```python
fm.Marker([37.86, -122.25], popup="You are here").add_to(my_map)
my_map
```
